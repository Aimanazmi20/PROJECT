# Import Data

```python
import pandas as pd

df = pd.read_csv('vgsales.csv')

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df.head()
```

#### -import vgsales data using pandas

# Clean Data

```python
df = df[df['Global_Sales'] >= 1.00]
df
```

#### - clean data using attribute global sales higher than 1.0

# Group by Platform

```python
platform_sales = df.groupby('Platform')['Global_Sales'].sum()
platform_sales = platform_sales.sort_values(ascending=False)

most_sales_platform = platform_sales.index[0]
most_sales_platform_value = platform_sales.iloc[0]

print(platform_sales)
print(most_sales_platform, most_sales_platform_value)
```

#### -total global sales for each platform

# Top 10 Game

```python
game_sales = df.sort_values('Global_Sales', ascending=False)
top_10_games = game_sales.head(10)

print(top_10_games)
```

#### - top 10 games with the highest global sales

# Most Sales by Year

```python
year_sales = df.groupby('Year')['Global_Sales'].sum()
year_sales = year_sales.sort_values(ascending=False)

most_sales_year = year_sales.index[0]
most_sale_year_value = year_sales.iloc[0]

print(year_sales)
print(most_sales_year, most_sale_year_value)
```

#### - total global sales for each years

# Most Sales by Genre

```python
genre_sales = df.groupby('Genre')['Global_Sales'].sum()
genre_sales = genre_sales.sort_values(ascending=False)

most_sales_genre = genre_sales.index[0]
most_sales_genre_value = genre_sales.iloc[0]

print(genre_sales)
print(most_sales_genre, most_sales_genre_value)
```

#### - most sales genre
#### from the result above, action genre is the most sales genre of all time follow by sports and shooter genre

# Most Sales by Publisher

```python
publiser_sales = df.groupby('Publisher')['Global_Sales'].sum()
publiser_sales = publiser_sales.sort_values(ascending=False)

most_sale_publisher = publiser_sales.index[0]
most_sale_publisher_value = publiser_sales.iloc[0]

print(publiser_sales)
print(most_sale_publisher, most_sale_publisher_value)
```

#### -publisher with the most sales
#### from the result above, nintendo has the highest sales of all other publisher

# Most Sales by Platform

```python
platform_game = df.groupby('Platform')['Global_Sales'].idxmax()
most_sale_platform_game = df.loc[platform_game, ['Name', 'Platform', 'Global_Sales']]

print(most_sale_platform_game)
```

#### - most sales game for each platform

# Most Sales Games in each Genre

```python
Genre_game = df.groupby('Genre')['Global_Sales'].idxmax()
most_sale_genre_game = df.loc[Genre_game, ['Name', 'Genre', 'Global_Sales']]

print(most_sale_genre_game)
```

### - most sales games for each genre

# Most Sales Game by Year

```python
Year_game = df.groupby('Year')['Global_Sales'].idxmax()
most_sale_year_game = df.loc[Year_game, ['Name', 'Year', 'Global_Sales']].sort_values('Year', ascending=False)

print(most_sale_year_game)
```

### most sales game for each year

# Most Sales Game by Publisher

```python
publisher_game = df.groupby('Publisher')['Global_Sales'].idxmax()
most_sale_publisher_game = df.loc[publisher_game, ['Name', 'Publisher', 'Global_Sales']]

print(most_sale_publisher_game)
```

### most sales game for each publisher

# Plotting Sales by Region in Each Year

```python
import matplotlib.pyplot as plt

plt.scatter(df['Year'], df['NA_Sales'], marker = "o", color = "r", alpha = 0.5, s = 5, label = ['NA Sales'])
plt.scatter(df['Year'], df['EU_Sales'], marker = "o", color = "g", alpha = 0.5, s = 5, label = ['EU Sales'])
plt.scatter(df['Year'], df['JP_Sales'], marker = "o", color = "b", alpha = 0.5, s = 5, label = ['JP Sales'])

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Year/Sales')

plt.legend(loc = 'upper left')

plt.show
```

### the scatterplot for NA, EU & JP sales by each year
### NA have the highest sales between EU & JP sales
